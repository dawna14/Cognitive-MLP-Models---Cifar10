```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('CNN_Model.xlsx')
```

```python
df.head()
```

```python
# summarize history for accuracy
plt.plot(df['Training Accuracy'])
plt.plot(df['Testing Accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(df['Loss'])
plt.plot(df['Val Loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
sns.pairplot(data=df[['Training Accuracy','Testing Accuracy']])
```

```python
sns.pairplot(data=df[['Loss','Val Loss']])
```

```python
colors = {'RMSprop': 'r', 'Adam': 'b', 'Adamax': 'g'}
df['Testing Accuracy'].plot(kind='bar', color=[colors[i] for i in df['Optimizer']])
plt.ylabel('Test Accuracy')
plt.xlabel('Optimizers')
plt.legend(labels=('RMSprop','Adam','Adamax'), loc='upper left',prop={'size': 10})
plt.rcParams['figure.figsize'] = (8, 6)
plt.show()
```
